Dans la table df_product (18 284 lignes) :

Colonne "StockCode"

Colonne "ProductCategory"
    - 2017 valeurs nulles

Colonne "ProductCategoryCode"
    - 1762 valeurs nulles
=> Les colonnes "ProductCategory" et "ProductCategoryCode" sont redondants : On va créer une colonne ProductCategory_Clean qui sera une combinaison des deux

Colonne "ProductFamily"
    - Aucune valeurs nulles
    - On va tout mettre en minuscule

Colonne "Description"
    - On na va pas retenir cette colonne pour notre analyse métier

Colonne "ProductNameClean"
    - 2512 valeurs nulles   => on va déterminer les valeurs nulles dans un premier temps avec le "stockCode" et dans un second temps avec le "UnitPriceGBP_Propre"

Colonne "UnitPriceGBP_Propre" (Conversion en nombres)
    - Aucune valeur nulle
    - 1 142 valeurs égale à 0  => Pour les valeurs égale à 0, on va estimer la valeur en utiliser la médiane du ratio "UnitPrice / UnitCost" et la valeur de "UnitCost"
    - 660 négatives     => On  va gérer les valeurs négatives avec des valeurs absolues

Colonne "UnitCostGBP_Propre" (conversion en nombre)
    - 812 valeurs nulles        => Pour les valeurs égale à 0, on va estimer la valeur en utiliser la médiane du ratio "UnitPrice / UnitCost" et la valeur de "UnitPrice"
    - 304 valeurs négatives     => Pour les valeurs négatives, on va prendre les valeurs absolues


```python
df_products=duckdb.query(f"""
-- Étape 1 : On crée un tableau virtuel avec tous nos calculs complexes
    SELECT DISTINCT
        StockCode,
        ProductCategory,
        ProductCategoryCode,
        ProductFamily,
        ProductNameClean, 
        REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
        REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre   
    FROM read_parquet('{chemin_fichier}')
""").df()

# ---------------------------------

# Création d'une table clients avec :
# ProductCategoryClean : soir la valeur de ProductCategory, soit la valeur de ProductCategoryCode si ProductCategory est null
# ProductFamilyClean : ProductFamily en minuscule
# ProductNameClean_2 : soit la valeur de ProductNameClean, soit la valeur max de ProductNameClean par StockCode si ProductNameClean est null

df_products_clean=duckdb.query(f"""
-- Étape 1 : On gère les conversions de base et les valeurs absolues (ABS)
    WITH Etape1_Nettoyage_Base AS (
        SELECT 
            TRIM(UPPER(StockCode)) AS StockCode,
            ABS(REPLACE(UnitPriceGBP, ',', '.')::FLOAT) AS Prix_Brut,
            ABS(REPLACE(UnitCostGBP, ',', '.')::FLOAT) AS Cout_Brut,
            TRIM(LOWER(COALESCE(ProductCategory, ProductCategoryCode))) AS ProductCategoryClean,
            TRIM(LOWER(ProductFamily)) as ProductFamilyClean,
            TRIM(LOWER(ProductNameClean)) AS Nom_Nettoye
        FROM read_parquet('{chemin_fichier}')
    ),
    
    -- Étape 2 : On calcule la médiane et on fait la 1ère passe sur les noms (par StockCode)
    Etape2_Calculs_Intermediaires AS (
        SELECT 
            *,
            MEDIAN(Prix_Brut / NULLIF(Cout_Brut, 0)) OVER () AS Ratio_Mediane,
            COALESCE(
                Nom_Nettoye, 
                MAX(Nom_Nettoye) OVER (PARTITION BY StockCode)
            ) AS Nom_Etape1
        FROM Etape1_Nettoyage_Base
    ),
    
    -- Étape 3 : On applique les formules d'imputation
    Etape3_Application_Regles AS (
        SELECT 
            StockCode,
            CASE
                WHEN ProductCategoryClean IS NULL THEN 'other'
                ELSE ProductCategoryClean
            END AS ProductCategory_Propre,
            ProductFamilyClean,
            COALESCE(
                Nom_Etape1,
                MAX(Nom_Etape1) OVER (PARTITION BY Prix_Brut),
                'inconnu'
            ) AS ProductNameClean_Final,
            CASE 
                WHEN Prix_Brut IS NULL OR Prix_Brut = 0 THEN Cout_Brut * Ratio_Mediane
                ELSE Prix_Brut 
            END AS UnitPriceGBP_Propre,
            CASE 
                WHEN Cout_Brut IS NULL OR Cout_Brut = 0 THEN Prix_Brut / NULLIF(Ratio_Mediane, 0)
                ELSE Cout_Brut 
            END AS UnitCostGBP_Propre
        FROM Etape2_Calculs_Intermediaires
    ),
    
    -- Étape 4 : Déduplication (On garde une seule ligne par StockCode avec la valeur la plus fréquente)
    Etape4_Deduplication AS (
        SELECT 
            StockCode,
            MODE(ProductCategory_Propre) AS ProductCategory_Propre,
            MODE(ProductFamilyClean) AS ProductFamilyClean,
            MODE(ProductNameClean_Final) AS ProductNameClean_Final,
            MODE(UnitPriceGBP_Propre) AS UnitPriceGBP_Propre,
            MODE(UnitCostGBP_Propre) AS UnitCostGBP_Propre
        FROM Etape3_Application_Regles
        GROUP BY StockCode
    )
    
    -- Résultat final
    SELECT * FROM Etape4_Deduplication
""").df()

# Requête initiale sans corrections 

# df_clients=duckdb.query(f"""
#     select distinct
#         StockCode,
#         ProductCategoryCode,
#         ProductCategory
#         ProductFamily,
#         ProductNameClean            
#     From read_parquet('{chemin_fichier}')
#     WHERE StockCode='22114'
# """).df()
```

Pour la table de faits, on va garder :

La colonne "OrderLineID"
    - Aucune valeur nulle

La colonne "Invoice"
    - Aucune valeur nulle

La colonne "Quantity":
    - Aucune valeure nulle
    - Mais des valeurs avec du texte : on va juste retirer les "units"
    - Des valeurs "négatifs" => On va considérer que ce sont des retours
    - Il y a 632 lignes de commandes avec la valeur "99999" => on va les remplacer par la médiane arrondie

La colonne "CustomerID"
    - 4416 valeurs nulles => On va les remplacer par 0 pour rester en format integer

La colonne "Country"
    - Aucune valeur nulle mais on va devoir corriger certaines valeurs de texte pour homogénéiser les noms de pays

La colonne "StockCode"
    - Aucune valeure nulle
    - On va rester en format texte

La colonne  "Promotion Code"
    - 45 valeurs nulles

```python
df_order_clean=duckdb.query(f"""
-- Étape 1 : On nettoie avec vos règles exactes
    WITH Etape1_Nettoyage AS (
        SELECT DISTINCT
            OrderLineID,
            TRIM(LOWER(OrderStatus)) AS OrderStatus,
            Invoice,         
            CASE
                WHEN CustomerID IS NULL THEN '0'
                ELSE CustomerID
            END AS CustomerID_clean,
            
            -- Homogénéisation des noms de pays
            CASE
                WHEN UPPER(TRIM(Country)) IN ('UNITED KINGDOM', 'UK', 'U.K.', 'GREAT BRITAIN') THEN 'United Kingdom'
                WHEN UPPER(TRIM(Country)) IN ('GERMANY', 'DEUTSCHLAND', 'DE')                  THEN 'Germany'
                WHEN UPPER(TRIM(Country)) IN ('FRANCE', 'FR')                                  THEN 'France'
                WHEN UPPER(TRIM(Country)) IN ('EIRE', 'REPUBLIC OF IRELAND')                   THEN 'Ireland'
                WHEN UPPER(TRIM(Country)) IN ('NETHERLANDS', 'HOLLAND', 'NL')                  THEN 'Netherlands'
                WHEN UPPER(TRIM(Country)) IN ('SWITZERLAND', 'CH')                             THEN 'Switzerland'
                WHEN UPPER(TRIM(Country)) IN ('AUSTRALIA', 'AUS')                              THEN 'Australia'
                WHEN UPPER(TRIM(Country)) IN ('USA', 'UNITED STATES')                          THEN 'United States'
                WHEN UPPER(TRIM(Country)) = 'UNITED ARAB EMIRATES'                             THEN 'United Arab Emirates'
                WHEN UPPER(TRIM(Country)) = 'SOUTH AFRICA'                                     THEN 'South Africa'
                ELSE UPPER(SUBSTRING(TRIM(Country), 1, 1)) || LOWER(SUBSTRING(TRIM(Country), 2))
            END AS Country_clean,
            
            TRIM(UPPER(StockCode)) AS StockCode,
            ABS(TRIM(REPLACE(Quantity, 'units', ''))::integer) as Quantity_clean,
            
            -- NOUVEAU : Déduction du PromotionCode grâce au DiscountRate
            CASE 
                WHEN TRY_CAST(DiscountRate AS FLOAT) = 0.05 THEN 'SPRING'
                WHEN TRY_CAST(DiscountRate AS FLOAT) = 0.10 THEN 'WELCOME10'
                WHEN TRY_CAST(DiscountRate AS FLOAT) = 0.15 THEN 'XMAS'
                WHEN TRY_CAST(DiscountRate AS FLOAT) = 0.20 THEN 'CLEARANCE'
                WHEN TRY_CAST(DiscountRate AS FLOAT) = 0.30 THEN 'VIP'
                WHEN TRY_CAST(DiscountRate AS FLOAT) = 0.00 THEN 'NONE'
                
                -- Si DiscountRate vaut 'None' ou est vide (TRY_CAST échoue), on garde le PromotionCode d'origine nettoyé
                ELSE UPPER(TRIM(COALESCE(PromotionCode, 'NONE')))
            END AS PromotionCode_clean
            
        FROM read_parquet('{chemin_fichier}')
    ),
    
    -- Étape 2 : Calcul des statistiques (Moyenne, Écart-Type, Médiane)
    Etape2_Stats AS (
        SELECT 
            *,
            -- 1. Calcul de la moyenne globale
            AVG(Quantity_clean) OVER () AS Moyenne_Globale,
            
            -- 2. Calcul de l'écart-type global
            STDDEV(Quantity_clean) OVER () AS Ecart_Type_Global,
            
            -- 3. Calcul de la médiane par produit (pour le remplacement)
            ROUND(MEDIAN(Quantity_clean) OVER (PARTITION BY StockCode), 0)::INTEGER AS Mediane_Quantite_Produit,
            
            -- 4. Médiane globale de secours (si le produit n'a pas d'historique valide)
            ROUND(MEDIAN(Quantity_clean) OVER (), 0)::INTEGER AS Mediane_Quantite_Globale
            
        FROM Etape1_Nettoyage
    )
    
    -- Étape 3 : On applique la règle des "3 Sigmas" sur la quantité
    SELECT 
        OrderLineID,
        CASE
            WHEN OrderStatus IS NULL THEN 'completed'
            ELSE OrderStatus
        END AS OrderStatus,
        Invoice,
        CustomerID_clean,
        Country_clean,
        StockCode,
        
        -- Si Quantité > Moyenne + (3 * Ecart-Type), on remplace par la médiane
        CASE 
            WHEN Quantity_clean > (Moyenne_Globale + (3 * Ecart_Type_Global)) 
                 THEN COALESCE(Mediane_Quantite_Produit, Mediane_Quantite_Globale)
            ELSE Quantity_clean
        END AS Quantity_clean,
        
        -- On affiche uniquement notre code promo final ultra-propre !
        PromotionCode_clean
        
    FROM Etape2_Stats
""").df()
```

```python
df_order_final = duckdb.query(f"""
    SELECT 
        -- 1. On sélectionne toutes les colonnes de notre table principale (Commandes)
        o.*,
        
        -- 2. On sélectionne UNIQUEMENT les colonnes voulues de la table secondaire (Produits)
        p.ProductNameClean_Final,
        p.ProductFamilyClean,
        o.Quantity_clean * p.UnitPriceGBP_Propre AS TotalPriceGBP,
    -- On définit nos tables et on leur donne des "surnoms" (alias) pour alléger le code
    FROM df_order_clean AS o
    
    -- 3. La clé de voûte : on relie les deux tables grâce au StockCode
    LEFT JOIN df_products_clean AS p
        ON o.StockCode = p.StockCode
""").df()

display(df_order_final.head())
```
