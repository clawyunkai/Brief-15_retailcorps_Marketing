Dans la table df_product (18 284 lignes) :

Colonne "StockCode"
    - Aucune valeure nulle

Colonne "ProductCategory"
    - 2017 valeurs nulles

Colonne "ProductCategoryCode"
    - 1762 valeurs nulles
=> Les colonnes "ProductCategory" et "ProductCategoryCode" sont redondants : On va créer une colonne ProductCategory_Clean qui sera une combinaison des deux

Colonne "ProductFamily"
    - Aucune valeurs nulles
    - On va tout mettre en minuscule

Colonne "Description"
    - On na va pas retenir cette colonne pour notre analyse métier

Colonne "ProductNameClean"
    - 2512 valeurs nulles   => on va déterminer les valeurs nulles dans un premier temps avec le "stockCode" et dans un second temps avec le "UnitPriceGBP_Propre"

Colonne "UnitPriceGBP_Propre" (Conversion en nombres)
    - Aucune valeur nulle
    - 1 142 valeurs égale à 0  => Pour les valeurs égale à 0, on va estimer la valeur en utiliser la médiane du ratio "UnitPrice / UnitCost" et la valeur de "UnitCost"
    - 660 négatives     => On  va gérer les valeurs négatives avec des valeurs absolues

Colonne "UnitCostGBP_Propre" (conversion en nombre)
    - 812 valeurs nulles        => Pour les valeurs égale à 0, on va estimer la valeur en utiliser la médiane du ratio "UnitPrice / UnitCost" et la valeur de "UnitPrice"
    - 304 valeurs négatives     => Pour les valeurs négatives, on va prendre les valeurs absolues


```python
df_products=duckdb.query(f"""
-- Étape 1 : On crée un tableau virtuel avec tous nos calculs complexes
    SELECT DISTINCT
        StockCode,
        ProductCategory,
        ProductCategoryCode,
        ProductFamily,
        ProductNameClean, 
        REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
        REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre   
    FROM read_parquet('{chemin_fichier}')
""").df()

# ---------------------------------

# Création d'une table clients avec :
# ProductCategoryClean : soir la valeur de ProductCategory, soit la valeur de ProductCategoryCode si ProductCategory est null
# ProductFamilyClean : ProductFamily en minuscule
# ProductNameClean_2 : soit la valeur de ProductNameClean, soit la valeur max de ProductNameClean par StockCode si ProductNameClean est null

df_products_clean=duckdb.query(f"""
-- Étape 1 : On gère les conversions de base et les valeurs absolues (ABS)
    WITH Etape1_Nettoyage_Base AS (
        SELECT 
            *,
            -- Conversion en FLOAT et mise en valeur absolue pour régler les 660 et 304 négatifs
            ABS(REPLACE(UnitPriceGBP, ',', '.')::FLOAT) AS Prix_Brut,
            ABS(REPLACE(UnitCostGBP, ',', '.')::FLOAT) AS Cout_Brut,
            
            -- Nettoyage des textes
            TRIM(LOWER(COALESCE(ProductCategory, ProductCategoryCode))) AS ProductCategoryClean,
            TRIM(LOWER(ProductFamily)) as ProductFamilyClean,
            
            -- MODIFICATION ICI : On applique TRIM et LOWER dès le début sur le nom
            TRIM(LOWER(ProductNameClean)) AS Nom_Nettoye
        FROM read_parquet('{chemin_fichier}')
    ),
    
    -- Étape 2 : On calcule la médiane et on fait la 1ère passe sur les noms (par StockCode)
    Etape2_Calculs_Intermediaires AS (
        SELECT 
            *,
            -- Calcul de la médiane du ratio (Prix / Coût)
            MEDIAN(Prix_Brut / NULLIF(Cout_Brut, 0)) OVER () AS Ratio_Mediane,
            
            -- Remplacement des noms vides grâce au StockCode
            COALESCE(
                Nom_Nettoye, 
                MAX(Nom_Nettoye) OVER (PARTITION BY StockCode)
            ) AS Nom_Etape1
        FROM Etape1_Nettoyage_Base
    ),
    
    -- Étape 3 : On applique les formules d'imputation finales
    Etape3_Application_Regles AS (
        SELECT DISTINCT
            StockCode,
            CASE
                WHEN ProductCategoryClean IS NULL THEN 'other'
                ELSE ProductCategoryClean
            END AS ProductCategory_Propre,
            ProductFamilyClean,
            
            -- MODIFICATION ICI : On ajoute 'inconnu' comme ultime solution de secours
            COALESCE(
                Nom_Etape1,
                MAX(Nom_Etape1) OVER (PARTITION BY Prix_Brut),
                'inconnu'
            ) AS ProductNameClean_Final,
            
            -- Imputation du Prix
            CASE 
                WHEN Prix_Brut IS NULL OR Prix_Brut = 0 THEN Cout_Brut * Ratio_Mediane
                ELSE Prix_Brut 
            END AS UnitPriceGBP_Propre,
            
            -- Imputation du Coût
            CASE 
                WHEN Cout_Brut IS NULL OR Cout_Brut = 0 THEN Prix_Brut / NULLIF(Ratio_Mediane, 0)
                ELSE Cout_Brut 
            END AS UnitCostGBP_Propre

        FROM Etape2_Calculs_Intermediaires
    )
    
    -- Résultat final
    SELECT * FROM Etape3_Application_Regles
""").df()

# Requête initiale sans corrections 

# df_clients=duckdb.query(f"""
#     select distinct
#         StockCode,
#         ProductCategoryCode,
#         ProductCategory
#         ProductFamily,
#         ProductNameClean            
#     From read_parquet('{chemin_fichier}')
#     WHERE StockCode='22114'
# """).df()
```

Pour la table de faits, on va garder :

La colonne "OrderLineID"
    - Aucune valeur nulle

La colonne "Invoice"
    - Aucune valeur nulle

La colonne "Quantity":
    - Aucune valeure nulle
    - Mais des valeurs avec du texte : on va juste retirer les "units"
    - Des valeurs "négatifs" => On va considérer que ce sont des retours

La colonne "CustomerID"
    - 4416 valeurs nulles => On va les remplacer par 0 pour rester en format integer

La colonne "Country"
    - Aucune valeur nulle mais on va devoir corriger certaines valeurs de texte pour homogénéiser les noms de pays

La colonne "StockCode"
    - Aucune valeure nulle
    - On va rester en format texte

La colonne  "Promotion Code"
    - 45 valeurs nulles

```python
df_order=duckdb.query(f"""
    SELECT DISTINCT
        OrderLineID,
        Invoice,
        TRIM(REPLACE(Quantity, 'units', ''))::integer as Quantity_clean,              
        CASE
            WHEN CustomerID IS NULL THEN '0'
            ELSE CustomerID
        END AS CustomerID_clean,
        Country,
        StockCode,           
        PromotionCode      
    FROM read_parquet('{chemin_fichier}')
""").df()


# On prend plutôt la colonne StockCode plutôt que ProductNameClean pour identifier les produits
# On enlève le DeliveryTime qui apporte peu d'informations
# On enlève le warehouseID qui est un identifiant de l'entrepôt et qui n'apporte pas d'informations sur le produit

df_order_clean=duckdb.query(f"""
-- Étape 1 : On nettoie les virgules et on convertit en nombres
    WITH Etape1_Nettoyage AS (
        SELECT DISTINCT
            OrderLineID,
            Invoice,
            Quantity,
            CustomerID,
            Country,
            REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
            REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre,
            StockCode,
            PromotionCode            
        FROM read_parquet('{chemin_fichier}')
    ),
    
    -- Étape 2 : On calcule la médiane du Markup sur toute la base
    Etape2_Mediane AS (
        SELECT 
            *,
            -- NULLIF(..., 0) est une sécurité pour ne jamais diviser par zéro
            UnitPriceGBP_Propre / NULLIF(UnitCostGBP_Propre, 0) AS Markup_Initial,
            
            -- MEDIAN() OVER() calcule la médiane globale pour pouvoir l'utiliser partout
            MEDIAN(UnitPriceGBP_Propre / NULLIF(UnitCostGBP_Propre, 0)) OVER () AS Mediane_Markup
        FROM Etape1_Nettoyage
    )
    
    -- Étape 3 : On applique votre règle de remplacement
    SELECT 
        OrderLineID,
        Invoice,
        Quantity,
        CustomerID,
        Country,
        UnitPriceGBP_Propre,
        
        -- Si le coût est vide, on prend (Prix / Mediane)
        COALESCE(UnitCostGBP_Propre, UnitPriceGBP_Propre / Mediane_Markup) AS UnitCostGBP_Propre,
        
        -- Le Markup final sera donc soit le vrai, soit la médiane
        COALESCE(Markup_Initial, Mediane_Markup) AS Markup,
        
        StockCode,
        PromotionCode
    FROM Etape2_Mediane
""").df()
```
