```python
from numpy import select
import pandas as pd
import duckdb
pd.set_option('display.max_columns', None)

chemin_fichier = r"retailcorps.parquet"

df=duckdb.query(f"""
    select distinct *
    From read_parquet('{chemin_fichier}')
""").df()
```

```python
def explorer_dataset(dataset):
        print(f"Start {'-'*10} Pour ce dataset, la taille est de {dataset.shape} {'-'*10}")
        print('-'*30)
        display(dataset.head(3))
        print('-'*30)
        display(dataset.tail(3))
        print('-'*30)
        print(f"Présences de valeurs 'NaN'.")
        print(f"{dataset.isna().sum()}")
        display(dataset[dataset.isna().any(axis=1)])
        print('-'*30)
        print(f"Les doublons")
        print(f"{dataset.duplicated().sum()}")
        print('-'*30)
        print(f"Les types de données")
        print(f"{dataset.dtypes}")
        print('-'*30)
        print(f"Description")
        display(dataset.describe().T)
        print("FIN"+'-'*100)
```

```python
explorer_dataset(df)
```

```python
df_clients=duckdb.query(f"""
-- Étape 1 : On crée un tableau virtuel avec tous nos calculs complexes
    SELECT DISTINCT
        StockCode,
        ProductCategory,
        ProductCategoryCode,
        ProductFamily,
        ProductNameClean, 
        REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
        REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre                    
    FROM read_parquet('{chemin_fichier}')
    
""").df()

# Création d'une table clients avec :
# ProductCategoryClean : soir la valeur de ProductCategory, soit la valeur de ProductCategoryCode si ProductCategory est null
# ProductFamilyClean : ProductFamily en minuscule
# ProductNameClean_2 : soit la valeur de ProductNameClean, soit la valeur max de ProductNameClean par StockCode si ProductNameClean est null

df_clients_clean=duckdb.query(f"""
-- Étape 1 : On crée un tableau virtuel avec tous nos calculs complexes
    WITH Base_Produits AS (
        SELECT DISTINCT
            StockCode,
            LOWER(COALESCE(ProductCategory, ProductCategoryCode)) AS ProductCategoryClean,
            LOWER(ProductFamily) as ProductFamilyClean,
            COALESCE(
                ProductNameClean, 
                MAX(ProductNameClean) OVER (PARTITION BY StockCode)
            ) AS ProductNameClean_2,
            REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
            REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre                    
        FROM read_parquet('{chemin_fichier}')
    )
    
    -- Étape 2 : On filtre très simplement sur la colonne qu'on vient de créer
    SELECT * FROM Base_Produits
    WHERE ProductNameClean_2 IS NULL
""").df()

# Requête initiale sans corrections 

# df_clients=duckdb.query(f"""
#     select distinct
#         StockCode,
#         ProductCategoryCode,
#         ProductCategory
#         ProductFamily,
#         ProductNameClean            
#     From read_parquet('{chemin_fichier}')
#     WHERE StockCode='22114'
# """).df()
```

```python
df_order=duckdb.query(f"""

    SELECT DISTINCT
        OrderLineID,
        Invoice,
        InvoiceDate,
        Quantity,
        CustomerID,
        Country,
        REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
        REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre,
        StockCode,
        ProductNameClean,              
        PromotionCode,
        DeliveryTime,
        WarehouseID            
    FROM read_parquet('{chemin_fichier}')

""").df()


# On prend plutôt la colonne StockCode plutôt que ProductNameClean pour identifier les produits
# On enlève le DeliveryTime qui apporte peu d'informations
# On enlève le warehouseID qui est un identifiant de l'entrepôt et qui n'apporte pas d'informations sur le produit

df_order_clean=duckdb.query(f"""
-- Étape 1 : On nettoie les virgules et on convertit en nombres
    WITH Etape1_Nettoyage AS (
        SELECT DISTINCT
            OrderLineID,
            Invoice,
            Quantity,
            CustomerID,
            Country,
            REPLACE(UnitPriceGBP, ',', '.')::FLOAT AS UnitPriceGBP_Propre,
            REPLACE(UnitCostGBP, ',', '.')::FLOAT AS UnitCostGBP_Propre,
            StockCode,
            PromotionCode            
        FROM read_parquet('{chemin_fichier}')
    ),
    
    -- Étape 2 : On calcule la médiane du Markup sur toute la base
    Etape2_Mediane AS (
        SELECT 
            *,
            -- NULLIF(..., 0) est une sécurité pour ne jamais diviser par zéro
            UnitPriceGBP_Propre / NULLIF(UnitCostGBP_Propre, 0) AS Markup_Initial,
            
            -- MEDIAN() OVER() calcule la médiane globale pour pouvoir l'utiliser partout
            MEDIAN(UnitPriceGBP_Propre / NULLIF(UnitCostGBP_Propre, 0)) OVER () AS Mediane_Markup
        FROM Etape1_Nettoyage
    )
    
    -- Étape 3 : On applique votre règle de remplacement
    SELECT 
        OrderLineID,
        Invoice,
        Quantity,
        CustomerID,
        Country,
        UnitPriceGBP_Propre,
        
        -- Si le coût est vide, on prend (Prix / Mediane)
        COALESCE(UnitCostGBP_Propre, UnitPriceGBP_Propre / Mediane_Markup) AS UnitCostGBP_Propre,
        
        -- Le Markup final sera donc soit le vrai, soit la médiane
        COALESCE(Markup_Initial, Mediane_Markup) AS Markup,
        
        StockCode,
        PromotionCode
    FROM Etape2_Mediane
""").df()
```
